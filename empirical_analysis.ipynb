```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

```python
records = []
with open('performance80.txt', 'r') as f:
    for line in f:
        split = line.split()
        records.append([int(split[3][:-1]), int(split[5])])
records.sort()
```

```python
edges = []
time = []
sumt = 0
for i in range(len(records)):
    sumt += records[i][1]
    if i % 10 == 9:
        edges.append(records[i][0])
        time.append(sumt / 10)
        sumt = 0
```

```python
edges, time = zip(*sorted(zip(edges, time)))
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['axes.labelsize'] = 10
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(edges, time)
ax.set(xlabel = 'number of edges', ylabel = 'time, µs')
plt.show()
```

```python
print(edges)
print(time)
```

```python
records = []
with open('performance.txt', 'r') as f:
    for line in f:
        split = line.split()
        records.append([int(split[1][:-1]), int(split[3][:-1]), int(split[5])])
records.sort()
```

```python
size = []
edges = []
time = []
sumt = 0
for i in range(len(records)):
    sumt += records[i][2]
    if i % 10 == 9:
        size.append(records[i][0])
        edges.append(records[i][1])
        time.append(sumt / 10)
        sumt = 0
```

```python
size_edges = [size[i]*edges[i] for i in range(len(size))]
```

```python
size_edges, time, size, edges = zip(*sorted(zip(size_edges, time, size, edges)))
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['axes.labelsize'] = 10
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(size_edges, time, c='red')
ax.plot(size_edges, [0.00315*edges[i]*math.log(size[i]) for i in range(len(size))])
ax.set(xlabel = 'vertices * edges', ylabel = 'time, µs')
plt.show()
```

```python
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['axes.labelsize'] = 40
fig = plt.figure(figsize=(35,35))
ax = plt.axes(projection='3d')
ax.plot_trisurf(size, edges, time)
ax.set(xlabel = 'vertices', ylabel = 'edges', zlabel = 'time, µs')
plt.show()
```

```python
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['axes.labelsize'] = 40
fig = plt.figure(figsize=(35,35))
ax = plt.axes(projection='3d')
ax.plot_trisurf(size, edges, time)
ax.plot_trisurf(size, edges, [0.00315*edges[i]*math.log(size[i]) for i in range(len(size))])
ax.set(xlabel = 'vertices', ylabel = 'edges', zlabel = 'time, µs')
plt.show()
```

```python
print(size)
print(edges)
print(time)
```
